```python
# Mentioning the required libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# To read the image:
image = cv2.imread('lego1.png')

# To convert the image to grayscale:
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# To plot the grayscale image:
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title('Grayscale image')
```

```python
#  To convert the image into a black and white (two-color) image (Binarize the image):
_, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
```

```python
# To plot the binary image:
plt.imshow(binary, cmap='gray')
plt.title('Binary image')
plt.show()
```

```python
# to preform the morphological operations to remove noise:
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)
```

```python
# To plot the image after morphological opening:
plt.imshow(opening, cmap='gray')
plt.title('Image after morphological opening')
plt.show()
```

```python
# To preform morphological closing to fill in small holes:
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel, iterations=3)
```

```python
# To preform the image after morphological closing:
plt.imshow(closing, cmap='gray')
plt.title('Image after morphological closing')
plt.show()
```

```python
# To erode away small circles inside the objects:
erosion = cv2.erode(closing, kernel, iterations=3)
```

```python
# To plot the image after erosion:
plt.imshow(erosion, cmap='gray')
plt.title('Image after erosion')
plt.show()
```

```python
# To preform edge detection:
edges = cv2.Canny(erosion, 50, 100)
```

```python
# To plot the edges:
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.show()
```

```python
# To find the contours of the edges:
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# To plot the contours on the original image:
image_contours = cv2.drawContours(image, contours, -1, (0,0,255), 2)
plt.imshow(image_contours)
plt.title('Contours on original image')
plt.show()
```

```python
# To print the number of objects in the image:
print(f'Number of objects in the image: {len(contours)}')
```
